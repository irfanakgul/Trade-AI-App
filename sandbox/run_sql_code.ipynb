```python
query = """
CREATE SCHEMA IF NOT EXISTS "user";

CREATE TABLE IF NOT EXISTS "user"."REGISTERED_USER_DETAILS" (
    "ID" BIGSERIAL PRIMARY KEY,

    "USERNAME" VARCHAR(50) NOT NULL,
    "EMAIL" VARCHAR(255) NOT NULL,
    "PASSWORD_HASH" VARCHAR(255) NOT NULL,

    "FIRST_NAME" VARCHAR(100) NOT NULL,
    "LAST_NAME" VARCHAR(100) NOT NULL,
    "DATE_OF_BIRTH" DATE NOT NULL,
    "COUNTRY" VARCHAR(100) NOT NULL,
    "MOBILE_PHONE" VARCHAR(25) NOT NULL,
    "GENDER" VARCHAR(20) NOT NULL,

    "EXCHANGE_EXPERIENCE" VARCHAR(20) NOT NULL,
    "ESTIMATED_START_BUDGET" NUMERIC(18,2) NOT NULL,

    "RESPONSIBILITY_APPROVED" BOOLEAN NOT NULL DEFAULT FALSE,

    "DEVICE_ID" VARCHAR(255) NOT NULL,

    "IS_ACTIVE" VARCHAR(20) NOT NULL DEFAULT 'NEW_USER',
    "USER_TYPE" VARCHAR(20) NOT NULL DEFAULT 'CLIENT',

    "CREATED_AT" TIMESTAMP NOT NULL DEFAULT CURRENT_TIMESTAMP,
    "UPDATED_AT" TIMESTAMP NOT NULL DEFAULT CURRENT_TIMESTAMP,
    "LAST_LOGIN_AT" TIMESTAMP NULL,

    CONSTRAINT "CHK_AGE"
        CHECK ("DATE_OF_BIRTH" <= CURRENT_DATE - INTERVAL '18 years'),

    CONSTRAINT "CHK_GENDER"
        CHECK ("GENDER" IN ('MALE', 'FEMALE', 'OTHER', 'PREFER_NOT_TO_SAY')),

    CONSTRAINT "CHK_EXPERIENCE"
        CHECK ("EXCHANGE_EXPERIENCE" IN ('BEGINNER', 'INTERMEDIATE', 'GOOD', 'VERY_GOOD')),

    CONSTRAINT "CHK_IS_ACTIVE"
        CHECK ("IS_ACTIVE" IN ('NEW_USER', 'ACTIVE')),

    CONSTRAINT "CHK_USER_TYPE"
        CHECK ("USER_TYPE" IN ('CLIENT'))
);
"""

run_postgres_sql(query, fetch=False)

print("✅")
```
