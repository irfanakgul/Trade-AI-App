```python
for exc in ["ams", "bist", "nasdaq", "nyse", "crypto", "assets"]:
    sql_text = f"""
    DROP TABLE IF EXISTS prod."{exc}_watch_dataset";

CREATE TABLE prod."{exc}_watch_dataset" (
    "EXCHANGE"   VARCHAR(32)   NOT NULL,
    "SYMBOL"     VARCHAR(64)   NOT NULL,
    "TIMESTAMP"  TIMESTAMP     NOT NULL,
    "OPEN"       DOUBLE PRECISION,
    "HIGH"       DOUBLE PRECISION,
    "LOW"        DOUBLE PRECISION,
    "CLOSE"      DOUBLE PRECISION,
    "VOLUME"     DOUBLE PRECISION,
    "ROW_ID"     VARCHAR(128)  NOT NULL,
    "SOURCE"     VARCHAR(64)   NOT NULL,
    "CREATED_AT" TIMESTAMP     NOT NULL DEFAULT NOW()
);

CREATE UNIQUE INDEX IF NOT EXISTS ux_{exc}_watch_dataset_row_id
    ON prod."{exc}_watch_dataset" ("ROW_ID");

CREATE INDEX IF NOT EXISTS ix_{exc}_watch_dataset_symbol
    ON prod."{exc}_watch_dataset" ("SYMBOL");

CREATE INDEX IF NOT EXISTS ix_{exc}_watch_dataset_exchange_symbol
    ON prod."{exc}_watch_dataset" ("EXCHANGE", "SYMBOL");

CREATE INDEX IF NOT EXISTS ix_{exc}_watch_dataset_timestamp
    ON prod."{exc}_watch_dataset" ("TIMESTAMP");
    """
    r = run_postgres_sql(sql_text, fetch=False)
    print(exc, r)

print("✅")
```
